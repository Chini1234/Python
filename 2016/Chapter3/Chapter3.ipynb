```python
drinks = {
    "item1" : {"珍珠奶茶","紅茶"},
    "item2" : {"烏龍奶茶","烏龍茶"},
    "item3" : {"蜂蜜紅茶","紅茶"},
   }
```

```python
for name, contents in drinks.items():
    if "紅茶" in contents:
        print(name)
```

```python
for name, contents in drinks.items():
    if "紅茶" in contents and not ("珍珠奶茶" in contents):
        print(name)
```

```python
for name, contents in drinks.items():
    if contents & {"紅茶"} :
        print(name)
```

```python
for name, contents in drinks.items():
    if "紅茶" in contents and not contents & {"蜂蜜紅茶"} :
        print(name)
```
